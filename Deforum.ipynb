## Now import the music from MusicGen for combining

```python
def combine_audio(vidname, audname, outname, fps=15):
    import moviepy.editor as mpe

    # Load video clip
    my_clip = mpe.VideoFileClip(vidname)

    # Load audio clip
    audio_background = mpe.AudioFileClip(audname)

    # Trim video to match audio duration if video is longer
    if my_clip.duration > audio_background.duration:
        my_clip = my_clip.subclip(0, audio_background.duration)
    # Trim audio to match video duration if audio is longer
    elif audio_background.duration > my_clip.duration:
        audio_background = audio_background.subclip(0, my_clip.duration)

    # Set audio to video clip
    final_clip = my_clip.set_audio(audio_background)

    # Write the final video file
    final_clip.write_videofile(outname, fps=fps)
```
